```python
def multiply_basic(poly1, poly2):
    result=[0]*(len(poly1)+len(poly2)-1)
    i=0;
    j=0;
    for x in poly1:
        #print('i-->',i)
        for y in poly2:
            #print('y-->',y)
            #print('sum-->',i+j)
            result[i+j]=round(result[i+j]+(x*y),2)
            #print('j-->',j)
            #print(result[i+j])
            j=j+1
            
        i=i+1
        j=0
    return(result)
         
```

Correct solution for the problem

```python
def add(poly1, poly2):
    """Add two polynomials"""
    result = [0] * max(len(poly1), len(poly2))
    for i in range(len(result)):
        if i < len(poly1):
            result[i] += poly1[i]
        if i < len(poly2):
            result[i] += poly2[i]
            
    resultr=[round(num, 2) for num in result]
    return resultr
```

```python
def sub(poly1, poly2):
    """Add two polynomials"""
    result = [0] * max(len(poly1), len(poly2))
    for i in range(len(result)):
        if i < len(poly1):
            result[i] += poly1[i]
        if i < len(poly2):
            result[i] -= poly2[i]
    resultr=[round(num, 2) for num in result]
    return resultr
```

```python
def split(poly1, poly2):
    """Split each polynomial into two smaller polynomials"""
    mid = max(len(poly1), len(poly2)) // 2
    return  (poly1[:mid], poly1[mid:]), (poly2[:mid], poly2[mid:])
```

```python
split([1, 2, 3, 4], [0, 4, 3, 6, 7, 8, 2])
```

```python
def increase_exponent(poly, n):
    """Multiply poly1 by x^n"""
    return [0] * n + poly
```
